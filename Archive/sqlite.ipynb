```python
import sqlite3
import pandas as pd
  
# Connect to SQLite database
conn = sqlite3.connect('data/spotify.sqlite')
```

```python
# Load CSV data into Pandas DataFrame
stud_data = pd.read_csv('data/data.csv')
# Write the data to a sqlite table
stud_data.to_sql('spotify', conn, if_exists='replace', index=False)
```

```python
# Create a cursor object
cur = conn.cursor()
# Fetch and display result
for row in cur.execute('SELECT * FROM spotify'):
    print(row)
```
